# Imports

```python
import numpy as np
from numpy import asarray
from PIL import Image, ImageDraw
from skimage.io import imread
import matplotlib.pyplot as plt
from math import *
import time
```

# Functions

```python
def importImage(path):
    imageOpen = Image.open(path)
    greyScaleImage = imageOpen.convert("L")
    imageArray = np.array(greyScaleImage, dtype='float64')
    return imageArray


def createWeightedMedianFilter():
    array = [[1, 3, 1], [3, 5, 3], [1, 3, 1]]
    return array


def createMedianFilter():
    array = [[1, 1, 1], [1, 1, 1], [1, 1, 1]]
    return array


def create_and_sort_V(image, appliedFilter):
    vector = []
    for i in range(0, 3):
        for j in range(0, 3):
            counter = appliedFilter[i][j]
            while(counter > 0):
                vector.append(image[i][j])
                counter = counter-1
    vector.sort()
    return vector


def Calculatemedian(vector):
    arrayLength = len(vector)
    if arrayLength % 2 == 0:
        firstValue = vector[arrayLength/2]
        secondValue = firstValue+1
        median = (vector[firstValue] + vector[secondValue]) / 2
        return median

    if arrayLength % 2 > 0:
        return (vector[arrayLength//2])
    

def calculateMedianAndApply(image):
    processedImage = importImage(image)
    BarCodeWeightedMedian = processedImage.copy() 
    BarCodeMedian = processedImage.copy() 
    weightedMedianFilter = createWeightedMedianFilter()
    medianFilter = createMedianFilter()
    limitrow = (processedImage.shape[0])-1
    limitcolumn = (processedImage.shape[1])-1
    for i in range(1, limitrow):
        for j in range(1, limitcolumn):
                image3x3 = [[processedImage[i-1, j-1], processedImage[i-1, j], processedImage[i-1, j+1]],
                            [processedImage[i, j-1], processedImage[i, j],processedImage[i, j+1]],
                            [processedImage[i+1, j-1], processedImage[i+1, j], processedImage[i+1, j+1]]]
                vectorWeightedMedian = create_and_sort_V(image3x3, weightedMedianFilter)
                medianWeightedMedian = Calculatemedian(vectorWeightedMedian)
                vectorMedian = create_and_sort_V(image3x3, medianFilter)
                medianMedian = Calculatemedian(vectorMedian)
                BarCodeWeightedMedian[i][j] = medianWeightedMedian
                BarCodeMedian[i][j] = medianMedian
    fig, ax1 = plt.subplots(figsize=(10, 10))
    ax1.imshow(BarCodeWeightedMedian, cmap='gray')
    plt.savefig('F:/GUC/Computer Vision/Project 3/data/BarCodeWeightedMedian.jpg')
    fig, ax2 = plt.subplots(figsize=(10, 10))
    ax2.imshow(BarCodeMedian, cmap='gray')
    plt.savefig('F:/GUC/Computer Vision/Project 3/data/BarCodeMedian.jpg')
```

# Comparison

Its clear that "BarCodeWeightedMedian.jpg" is better than "BarCodeMedian.jpg" as the borders between the lines and the details are more preserved. The weighted median filter is better because it takes into consideration the line pattern in the barcode. This means that the target pixel have more impact on the line pattern than the other pixels. This is true in the barcode case as the center pixel is the pixels that contribute to the line's length and width the most. Since the normal median filter does not take that into consideration the result appears more blurry and less detailed. The difference in the weights between the diagonals and the above/below/side pixels effect appears when you take into consideration the noise. Because, if there is a noisy pixel it will have a smaller effect on the pixels as it will only be amplified once when being on the side. 

The lines are not the only thing in the images too. The effect of giving less weights to the diagonal pixels shines when we look at the numbers. As diagonals are completely left out in a number like 2. As its curvature means that the upper right diagonal pixel should not contribute (at its border for example). Thats why numbers are way clearer in "BarCodeWeightedMedian.jpg".
